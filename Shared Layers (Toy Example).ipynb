```python
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras import Model

import numpy as np
```

```python
# Wie definieren ein Modell: 
# 1. Den Input (eine nat√ºrliche Zahl)
```

```python
input_layer = Input(shape=(1,), name="Input_Layer")
```

```python
print(input_layer)
```

```python
input_for_training = [2,3,4,5,6,7]
results_1 = np.array(input_for_training) * 3
results_2 = [2] + np.array(input_for_training)
print (results_1, results_2)

# results_total = tf.concat([[results_1], [results_2]],0)

results_total = [[6,4],[9,5],[12,6],[15,7], [18,8], [21,9]]
print (results_total)
```

```python
shared_layer = Dense(2, name="shared_layer")(input_layer)
out_layer_1 = Dense(1, name ="out_1")(shared_layer)
out_layer_2 = Dense(1, name="out_2")(shared_layer)
total_out = tf.concat([out_layer_1, out_layer_2],1, name="concat_output")

model = Model(inputs=input_layer, outputs=total_out)

model.summary()

```

```python
model.compile(optimizer="SGD", loss='mae')

print(model.predict([1]))
```

```python
model.fit(input_for_training, results_total, epochs=50)
```

```python
model.predict([10])
```
